```python
from keras.models import Sequential
from keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

```python
from scikeras.wrappers import KerasClassifier
```

```python
model=Sequential()

model.add(Dense(units=20,input_dim=8,activation='relu',kernel_initializer='glorot_uniform'))

model.add(Dense(units=60,activation='relu',kernel_initializer='glorot_uniform'))

model.add(Dense(units=50,activation='relu',kernel_initializer='glorot_uniform'))

model.add(Dense(units=1,activation='sigmoid',kernel_initializer='glorot_uniform'))

model.summary()
```

```python
def create_model():

  model=Sequential()

  model.add(Dense(units=20,input_dim=8,activation='relu',kernel_initializer='glorot_uniform'))

  model.add(Dense(units=60,activation='relu',kernel_initializer='glorot_uniform'))

  model.add(Dense(units=50,activation='relu',kernel_initializer='glorot_uniform'))

  model.add(Dense(units=1,activation='sigmoid',kernel_initializer='glorot_uniform'))

  model.compile(optimizer=Adam(learning_rate=0.001),loss='binary_crossentropy',metrics=['accuracy'])

  model.summary()

  return model
```

```python
model =KerasClassifier(model=create_model,loss="binary_crossentropy", verbose=0)
```

```python
grid = GridSearchCV(
    estimator=model,
    param_grid={'batch_size': [10, 20], 'epochs': [10, 12]},
    cv=3)
```

```python
result=grid.fit(x_train,y_train)
```
